## Soru 1 (12 puan):
* Bir derse kaydolan öğrencilerin listesini içeren ekteki "Ogrenci Listesi.xls" dosyasını ```ogr_list``` adlı DataFrame'e okuyun.
* İlk sütunu silin.
* "Açıklama" sütunundaki ```Nan``` değerlerini ```İlk``` değeri ile, diğer değerleri ise ```Tekrar``` değeri ile değiştirin.
* "Açıklama" sütununun ismini "Tekrar Durumu" olacak şekilde değiştirin.
* Öğrenci numarasının 4. ve 5. karakterleri, öğrencinin okula giriş yılının son 2 basamağını temsil etmektedir. Buna göre DataFrame'e, öğrencilerin okula giriş yılı verilerini ```20##``` formatında içeren "Okula Giriş Yılı" isimli bir sütun ekleyin.
* "Öğrenci No" sütununu DataFrame'in index'i yapın.
* En son oluşan DataFrame'in satır ve sütun sayısını yazdırın.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)  

# Excel dosyasını DataFrame'e oku
ogr_list = pd.read_excel("Ogrenci Listesi.xls")

# İlk sütunu sil
ogr_list = ogr_list.drop(columns=["No"])

# Değerleri değiştirmek için lambda fonksiyonu tanımlanır
replace_values = lambda x: 'Tekrar' if pd.notnull(x) else 'İlk'

# Lambda fonksiyonu ilgili sütuna uygulanır
ogr_list["Açıklama"] = ogr_list["Açıklama"].map(replace_values)

# "Açıklama" sütununun ismini değiştir
ogr_list.rename(columns={"Açıklama": "Tekrar Durumu"}, inplace=True)

# "Okula Giriş Yılı" sütununu ekleyin
ogr_list["Okula Giriş Yılı"] = '20' + ogr_list["Öğrenci No"].str[3:5]

# "Öğrenci No" sütununu DataFrame'in index'i yap
ogr_list.set_index("Öğrenci No", inplace=True)


# Sonuçları yazdır
print("Satır Sayısı:", ogr_list.shape[0])
print("Sütun Sayısı:", ogr_list.shape[1])

ogr_list

````

## Soru 2 (12 puan):
* Öğrencilerin okula giriş yıllarının dersi tekrar durumlarına göre dağılımını (öğrenci sayılarını), yatay yığılmış (stacked) sütun grafik olarak gösterin.
* Derse tekrar kaydolan öğrencilerin, okula kayıt yıllarına göre dağılımlarını (yüzdelerini) pasta grafik olarak gösterin.

````python
# Öğrenci sayılarını hesapla
count_by_year_status = ogr_list.groupby(["Okula Giriş Yılı", "Tekrar Durumu"]).size().unstack(fill_value=0)
# Yatay yığılmış sütun grafiği oluştur
count_by_year_status.plot(kind="barh", stacked=True)
plt.xlabel("Öğrenci Sayısı")
plt.ylabel("Okula Giriş Yılı")
plt.title("Öğrenci Sayısı - Okula Giriş Yılı ve Tekrar Durumu")
plt.legend(title="Tekrar Durumu")
plt.show()
# Derse tekrar kaydolan öğrencilerin, okula kayıt yıllarına göre yüzdelerini hesapla
percentage_by_year = ogr_list[ogr_list["Tekrar Durumu"] == "Tekrar"].groupby("Okula Giriş Yılı")["Tekrar Durumu"].value_counts(normalize=False)

# Pasta grafiği oluştur
percentage_by_year.plot(kind="pie", subplots=True, figsize=(8, 4), autopct="%1.1f%%", legend=False)
plt.title("Derse Tekrar Kaydolan Öğrencilerin Okula Kayıt Yıllarına Göre Dağılımı")
plt.ylabel("")
plt.show()
````

## Soru 3 (20 puan):
* Bu derse ilk kez kaydolan öğrenciler, laboratuvar saatlerine girmek ve orada verilen ödevleri yapmak zorundadır. Derse tekrar kaydolan öğrenciler ise dönem başında laboratuvar saatlerine girmek isteyip istemediklerini bildirmektedirler ve girmek istiyorlarsa yine orada verilen ödevleri yapmak zorundadırlar. Buna göre dersin laboratuvar saatlerine giren öğrencilerin orada verilen ödevlerden aldıkları notları içeren ekteki "Lab Notlari.xls" dosyasını ```lab_notlari``` adlı DataFrame'e okuyun.
* İlk sütunu silin.
* "Öğrenci No" sütununu DataFrame'in index'i yapın.
* "Lab-* Notu" sütunlarındaki eksik veriler, öğrencinin mazereti nedeniyle ilgili ödevden muaf olduğunu göstermektedir. Buna göre DataFrame'e, öğrencilerin mazeretli olmadıkları laboratuvar saatlerinde verilen ödevlerden aldıkları notların ortalamasını (en düşük 2 not dışarıda bırakılacaktır ve en yakın tamsayıya yuvarlatılacaktır) içeren "Lab Notu" isimli bir sütun ekleyin.
* En son oluşan DataFrame'in satır ve sütun sayısını yazdırın.
* Lab saatlerinde verilen ödevlerin (10 adet) sınıf not ortalamalarını gösteren bir sütun grafik oluşturun.

````python
# Excel dosyasını okuyun ve "Lab Notlari" adlı bir DataFrame'e yükleyin
lab_notlari = pd.read_excel("Lab Notlari.xls")

# İlk sütunu sil
lab_notlari = lab_notlari.drop(columns=["No"])

# "Öğrenci No" sütununu DataFrame'in index'i yap
lab_notlari.set_index("Öğrenci No", inplace=True)


# "Lab Notu" isimli yeni bir sütun ekle ve ödev notlarının ortalamasını hesapla (en düşük 2 not alına)
orta_hesapla = lambda lab_notlari: (
    lab_notlari.assign(
        **{"Lab Notu": np.round(
            (lab_notlari.sum(axis=1) - np.partition(lab_notlari, 2, axis=1)[:, :2].sum(axis=1))
            / ((~np.isnan(lab_notlari)).sum(axis=1) - 2)
        )}
    )
)

lab_notlari = orta_hesapla(lab_notlari)

# Satır ve sütun sayısını yazdırın
print("Satır Sayısı:", lab_notlari.shape[0])
print("Sütun Sayısı:", lab_notlari.shape[1])

# Ödev notlarının sınıf ortalamalarını içeren bir sütun grafik oluştur
odev_ortalamalari = lab_notlari.iloc[:, 0:].mean().round(2)[:-1]
plt.bar(range(len(odev_ortalamalari)), odev_ortalamalari)
plt.xticks(range(len(odev_ortalamalari)), range(1, len(odev_ortalamalari)+1))  # Yeni satır
for i in range(len(odev_ortalamalari)):
    plt.text(i, odev_ortalamalari[i], str(odev_ortalamalari[i]), ha='center', va='bottom')
plt.xlabel("Ödevler")
plt.ylabel("Ortalama Not")
plt.title("Ödevlerin Sınıf Ortalamaları")
plt.show()

lab_notlari


````

## Soru 4 (20 puan):
* Bu derste öğrencilere 2 adet proje verilmektedir. Öğrenciler projeleri isterlerse tek başlarına, isterlerse 2 kişilik grup oluşturarak yapabilmektedir. Grup oluşturulurken öğrencilerin numaraları alt çizgi (_) karakteri ile birleştirilmektedir. Grup üyeleri projeden aynı notu almaktadır. Buna göre projeleri teslim eden öğrencilerin/grupların projelerden aldıkları notları içeren ekteki "1.Proje Notlari.xls" ve "2.Proje Notlari.xls" dosyalarını sırasıyla ```proje1_notlari``` ve ```proje2_notlari``` adlı DataFrame'lere okuyun.
* İkinci proje notu en yüksek ve en düşük olan öğrencilerin/grupların numaralarını yazdırın.
* ```proje1_notlari``` ve ```proje2_notlari``` adlı DataFrame'leri, index'i "Öğrenci No" olan ve projeyi teslim eden her öğrencinin ayrı ayrı notunun yer aldığı DataFrame'ler haline getirin. Bir başka deyişle, ```proje1_notlari``` DataFrame'inin index'i "Öğrenci No" olmalı ve "1. Proje Notu" isimli bir sütun içermelidir. Benzer şekilde, ```proje2_notlari``` DataFrame'inin index'i "Öğrenci No" olmalı ve "2. Proje Notu" isimli bir sütun içermelidir.
* En son oluşan DataFrame'lerin satır sayılarını yazdırın.

````python
# Excel dosyalarını DataFrame'lere oku
proje1_notlari = pd.read_excel("1.Proje Notlari.xls")
proje2_notlari = pd.read_excel("2.Proje Notlari.xls")

proje1_notlari = proje1_notlari.drop(columns=["No"])
proje2_notlari = proje2_notlari.drop(columns=["No"])

# İkinci proje notu en yüksek olan öğrenci/grupları bul
en_yuksek = proje2_notlari[proje2_notlari["2. Proje Notu"] == proje2_notlari["2. Proje Notu"].max()]["Öğrenci/Grup No "]

# İkinci proje notu en düşük olan öğrenci/grupları bul
en_dusuk = proje2_notlari[proje2_notlari["2. Proje Notu"] == proje2_notlari["2. Proje Notu"].min()]["Öğrenci/Grup No "]

print("Proje 2 için en yüksek notu alan öğrenci/grup(lar):")
print(en_yuksek.to_list())

print("Proje 2 için en düşük notu alan öğrenci/grup(lar):")
print(en_dusuk.to_list())

# 2 kişilik grupları ayırarak ayrı öğrenciler olarak kaydet
# 2'ye ayır
def split_group_number(row):
    or_no = str(row["Öğrenci/Grup No "])
    if '_' in or_no:
        ogrenciler = or_no.split('_')
        row["Öğrenci/Grup No "] = ogrenciler[0]
        new_row = row.copy()
        new_row["Öğrenci/Grup No "] = ogrenciler[1]
        return [row, new_row]
    else:
        return [row]
# Dataframe'i güncelle eğer grup durumu varsa ikinci kişiyi yeni bir satır olarak ekle
def update_dataframe(df):
    for index, row in df.iterrows():
        rows = split_group_number(row)
        df.loc[index] = rows[0]
        if len(rows)== 2:
             df = pd.concat([df, pd.DataFrame([rows[1]])], ignore_index=True)
    df.rename(columns={"Öğrenci/Grup No ": "Öğrenci No"}, inplace=True)
    df.set_index("Öğrenci No", inplace=True)
    return df

proje1_notlari = update_dataframe(proje1_notlari)
proje2_notlari = update_dataframe(proje2_notlari)
print(" ")
print("Satır Sayısı:", proje1_notlari.shape[0])
print("Sütun Sayısı:", proje1_notlari.shape[1])
print("Satır Sayısı:", proje2_notlari.shape[0])
print("Sütun Sayısı:", proje2_notlari.shape[1])

print(proje1_notlari)
proje2_notlari





````

## Soru 8 (24 puan):
* Derse kayıtlanan tüm öğrencilerin tüm notlarını içermesi için ```ogr_list``` adlı DataFrame'i "Öğrenci No" aracılığıyla diğer tüm DataFrame'ler ile birleştirin, bu sırada ```lab_notlari``` DataFrame'inden sadece "Lab Notu" sütununun alınmasını sağlayın.
* ```ogr_list``` DataFrame'nin "1. Proje Notu", "2. Proje Notu" ve "Arasınav Notu" sütunlarındaki ```Nan``` değerlerini ```0``` değeri ile değiştirin.
* Öğrencilerin bu derse ilişkin proje notu, 1. proje notunun 3/7'si ile 2. proje notunun 4/7'si toplanarak (en yakın tamsayıya yuvarlatılacaktır) bulunmaktadır. Buna göre DataFrame'e, "Proje Notu" isimli bir sütun ekleyin.
* Bu derse tekrar kaydolan öğrencilerden laboratuvar saatlerine girmek isteyen, dolayısıyla orada verilen ödevleri de yapan öğrencilerin numara, lab notu ve proje notu verilerini listeleyin.
* Bu derse tekrar kaydolan öğrencilerden laboratuvar saatlerine girmek istemeyen, dolayısıyla orada verilen ödevleri yapmayan öğrencilerin proje notları, aynı zamanda lab notu yerine de sayılmaktadır. Buna göre "Lab Notu" sütunundaki ```Nan``` değerlerini, ilgili öğrencilerin "Proje Notu" sütunundaki değerleri ile değiştirin.
* Öğrencilerin bu derse ilişkin dönem sonu notu, lab notunun %13'ü, proje notunun %22,75'i, ara sınav notunun %29,25'i ve final/bütünleme sınavı notunun %35'i toplanarak (en yakın tamsayıya yuvarlatılacaktır) bulunmaktadır. Ancak öğrenci final ve bütünleme sınavlarına girmediyse, dönem sonu notu doğrudan 0 olmaktadır. Ayrıca öğrenci hem final hem bütünleme sınavına girdiyse, final sınavı notu geçerliliğini kaybetmektedir. Buna göre DataFrame'e, "Dönem Sonu Notu" isimli bir sütun ekleyin.

````python
# Birleştirme işlemi
ogr_list = ogr_list.merge(lab_notlari[["Lab Notu"]], on="Öğrenci No", how="outer")
ogr_list = ogr_list.merge(proje1_notlari, on="Öğrenci No", how="outer")
ogr_list = ogr_list.merge(proje2_notlari, on="Öğrenci No", how="outer")
ogr_list = ogr_list.merge(vize_notlari, on="Öğrenci No", how="outer")
ogr_list = ogr_list.merge(final_notlari, on="Öğrenci No", how="outer")
ogr_list = ogr_list.merge(butunleme_notlari, on="Öğrenci No", how="outer")
#NaN'ları 0 yap 
ogr_list["1. Proje Notu"].fillna(0, inplace=True)
ogr_list["2. Proje Notu"].fillna(0, inplace=True)
ogr_list["Arasınav Notu"].fillna(0, inplace=True)
# Proje Notu hesaplama fonksiyonu
def hesapla(row):
    proje1 = row["1. Proje Notu"]
    proje2 = row["2. Proje Notu"]
    proje_notu = round((proje1 * 3/7) + (proje2 * 4/7))
    return proje_notu

# "Proje Notu" sütununu oluşturma
ogr_list["Proje Notu"] = ogr_list.apply(hesapla, axis=1)
# Laba giren tekrarcıları yazdır
laba_giren_tekrarcilar = ogr_list.loc[(ogr_list["Tekrar Durumu"] == "Tekrar") & (ogr_list["Lab Notu"].notna())]
print("Tekrar durumunda olup laba girenler")
print(laba_giren_tekrarcilar[["Lab Notu", "Proje Notu"]])
#Laba girmeyen tekrarcıların Proje Notlarını Lab Notu Yap
ogr_list.loc[(ogr_list['Tekrar Durumu'] == 'Tekrar') & (ogr_list['Lab Notu'].isnull()), 'Lab Notu'] = ogr_list.loc[(ogr_list['Tekrar Durumu'] == 'Tekrar') & (ogr_list['Lab Notu'].isnull()), 'Proje Notu']

# Dönem Sonu Notu hesaplama fonksiyonu
def donem_hesapla(row):
    lab_notu = row["Lab Notu"]
    proje_notu = row["Proje Notu"]
    arasinav_notu = row["Arasınav Notu"]
    final_notu = row["Final Sınavı Notu"]
    butunleme_notu = row["Bütünleme Sınavı Notu"]

    if pd.isna(final_notu) and pd.isna(butunleme_notu):
        donem_sonu_notu = 0
    elif pd.notna(butunleme_notu):
        donem_sonu_notu = round((lab_notu * 0.13) + (proje_notu * 0.2275) + (arasinav_notu * 0.2925) + (butunleme_notu * 0.35))
    else:
        donem_sonu_notu = round((lab_notu * 0.13) + (proje_notu * 0.2275) + (arasinav_notu * 0.2925) + (final_notu * 0.35))

    return donem_sonu_notu

# "Dönem Sonu Notu" sütununu oluşturma
ogr_list["Dönem Sonu Notu"] = ogr_list.apply(donem_hesapla, axis=1)  
ogr_list
````

## Soru 9 (13 puan):
